# BoAmps — JSON to Parquet pipeline

Reads all `.json` files from a source folder and writes one Parquet file per table into `db_model/parquet/`.

```python
import logging
import sys
from pathlib import Path

from huggingface_hub import snapshot_download

REPO_ROOT = Path().resolve().parents[1]
if str(REPO_ROOT) not in sys.path:
    sys.path.insert(0, str(REPO_ROOT))

from db_model.constants import PARQUET_DIR
from db_model.insertors.build_parquet import build
from db_model.retrievers.algorithms import get_algorithms
from db_model.retrievers.datasets import get_datasets
from db_model.retrievers.infrastructure import get_infrastructure
from db_model.retrievers.measures import get_measures
from db_model.retrievers.reports import get_reports, get_reports_cross_table
from methods.visualisations.algorithms import plot_algorithms
from methods.visualisations.datasets import plot_datasets
from methods.visualisations.infrastructure import plot_infrastructure
from methods.visualisations.measures import plot_measures
from methods.visualisations.reports import plot_reports, plot_reports_cross_table, plot_reports_over_time
```

```python
HF_REPO = "boavizta/open_data_boamps"

# Downloads (or reuses cached) dataset repo
hf_local = Path(snapshot_download(repo_id=HF_REPO, repo_type="dataset"))
SOURCE_DIR = hf_local / "data"

print(f"Source : {SOURCE_DIR}")
print(f"Output : {PARQUET_DIR}")
```

```python
logging.basicConfig(level=logging.INFO)

build(data_dir=SOURCE_DIR)

print("\nParquet files written:")
for f in sorted(PARQUET_DIR.glob("*.parquet")):
    print(f"  {f.name:30s}  {f.stat().st_size / 1024:.1f} KB")
```
